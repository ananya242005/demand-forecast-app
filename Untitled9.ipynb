```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from prophet import Prophet
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
import pandas as pd

df = pd.read_csv("ML.txt")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns = ['ds', 'y']
df['ds'] = pd.to_datetime(df['ds'])
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(df['ds'], df['y'])
plt.title("Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
from prophet import Prophet

model = Prophet()
model.fit(df)
```

```python
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

forecast.head()
```

```python
model.plot(forecast)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Take only original date range (ignore future 30 days)
forecast_train = forecast[:len(df)]

mae = mean_absolute_error(df['y'], forecast_train['yhat'])
rmse = np.sqrt(mean_squared_error(df['y'], forecast_train['yhat']))

print("MAE:", mae)
print("RMSE:", rmse)
```

```python
train_size = int(len(df) * 0.8)

train = df[:train_size]
test = df[train_size:]

print("Train size:", len(train))
print("Test size:", len(test))
```

```python
model = Prophet()
model.fit(train)
```

```python
future = model.make_future_dataframe(periods=len(test))
forecast = model.predict(future)

forecast.tail()
```

```python
forecast_test = forecast[-len(test):]

mae = mean_absolute_error(test['y'], forecast_test['yhat'])
rmse = np.sqrt(mean_squared_error(test['y'], forecast_test['yhat']))

print("Test MAE:", mae)
print("Test RMSE:", rmse)
```

```python
plt.figure(figsize=(12,6))

plt.plot(train['ds'], train['y'], label="Train")
plt.plot(test['ds'], test['y'], label="Actual Test")
plt.plot(test['ds'], forecast_test['yhat'], label="Predicted Test")

plt.legend()
plt.show()
```
